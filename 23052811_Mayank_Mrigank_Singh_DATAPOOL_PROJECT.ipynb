```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

file_path ='/content/sample_data/global_superstore_2016 (1).xlsx'

df = pd.read_excel(file_path)

```

 1) Max. Shipping  cost?

```python
max_shipping_cost = df['Shipping Cost'].max()

# Display the result
print(f"The maximum shipping cost is: {max_shipping_cost}")
```

2) What are the 3 countries that generated the highest total profit for Global Superstore?


```python
# Group by 'Country' and sum the 'Profit'
country_profit = df.groupby('Country')['Profit'].sum().reset_index()

# Sort by 'Profit' in descending order and get the top 3 countries
top_3_countries = country_profit.sort_values(by='Profit', ascending=False).head(3)

# Display the result
print("Top 3 countries by total profit:")
print(top_3_countries)
```

3)Identify the category with the highest average sales  in the United States


```python
# Filter the dataset for the United States
us_data = df[df['Country'] == 'United States']

# Group by 'Category' and calculate the average 'Sales'
category_avg_sales = us_data.groupby('Category')['Sales'].mean().reset_index()

# Sort by 'Sales' in descending order and get the top category
top_category = category_avg_sales.sort_values(by='Sales', ascending=False).head(1)

# Display the result
print("Category with the highest average sales in the United States:")
print(top_category)
```

4)Access Afghanistan  profitability (i.e., total profit) ,How does it compare to  other       Asian countries?


```python
# List of Asian countries (including Afghanistan)
asian_countries = [
    'Afghanistan', 'China', 'India', 'Indonesia', 'Japan', 'Malaysia',
    'Pakistan', 'Philippines', 'Singapore', 'South Korea', 'Thailand',
    'Vietnam', 'Bangladesh', 'Sri Lanka', 'Nepal'
]

# Filter data for Afghanistan
afghanistan_data = df[df['Country'] == 'Afghanistan']

# Filter data for other Asian countries
asian_data = df[df['Country'].isin(asian_countries)]

# Calculate total profit for Afghanistan
afghanistan_profit = afghanistan_data['Profit'].sum()

# Calculate total profit for other Asian countries (excluding Afghanistan)
asian_profit = asian_data.groupby('Country')['Profit'].sum().reset_index()

# Display Afghanistan's total profit
print(f"Total profit for Afghanistan: {afghanistan_profit}")

# Display total profit for other Asian countries
print("Total profit for other Asian countries:")
print(asian_profit)

# Compare Afghanistan's profit to the other Asian countries
asian_comparison = asian_profit[asian_profit['Country'] != 'Afghanistan']
print("Comparison of Afghanistan's profit to other Asian countries:")
print(asian_comparison)
```

6) Which product subcategory has the highest average profit in Australia?


```python
# Filter the dataset for Australia
australia_data = df[df['Country'] == 'Australia']

# Group by 'Sub-Category' and calculate the average 'Profit'
subcategory_avg_profit = australia_data.groupby('Sub-Category')['Profit'].mean().reset_index()

# Sort by 'Profit' in descending order and get the top subcategory
top_subcategory = subcategory_avg_profit.sort_values(by='Profit', ascending=False).head(1)

# Display the result
print("Product subcategory with the highest average profit in Australia:")
print(top_subcategory)
```

7) Who are the most valuable customers and what do they purchase?

```python
# Group by 'Customer ID' and sum the 'Profit' or 'Sales'
customer_value = df.groupby('Customer ID')['Profit'].sum().reset_index()

# Sort to find the most valuable customers
most_valuable_customers = customer_value.sort_values(by='Profit', ascending=False).head(10)  # Get top 10 customers

# Merge with the original DataFrame to find their purchases
valuable_customers_purchases = pd.merge(most_valuable_customers, df, on='Customer ID', how='inner')

# Select relevant columns to display
result = valuable_customers_purchases[['Customer ID', 'Profit_x', 'Product Name', 'Quantity', 'Sales']]

# Display the most valuable customers and what they purchase
print("Most Valuable Customers and their Purchases:")
print(result)
```

8) What is the count of product name in Australia?




```python
# Filter the dataset for Australia
australia_data = df[df['Country'] == 'Australia']

# Count the occurrences of each product name
product_counts = australia_data['Product Name'].value_counts()

# Display the count of product names
print("Count of each product name in Australia:")
print(product_counts)
```

9)What  is the situation of discount in each type of products in India?



```python
# Filter the dataset for India
india_data = df[df['Country'] == 'India']

# Group by 'Category' (or 'Sub-Category' if you have that information) and calculate average discount
discount_analysis = india_data.groupby('Category').agg(
    average_discount=('Discount', 'mean'),
    total_products_with_discount=('Discount', lambda x: (x > 0).sum())
).reset_index()

# Sort by average discount in descending order
discount_analysis = discount_analysis.sort_values(by='average_discount', ascending=False)

# Display the results
print("Discount situation for each type of products in India:")
print(discount_analysis)
```

10) What is the category wise sum of sales of Indonesia?

```python
# Filter the dataset for Indonesia
indonesia_data = df[df['Country'] == 'Indonesia']

# Group by 'Category' and calculate the sum of 'Sales'
category_sales_sum = indonesia_data.groupby('Category')['Sales'].sum().reset_index()

# Sort by 'Sales' in descending order
category_sales_sum = category_sales_sum.sort_values(by='Sales', ascending=False)

# Display the results
print("Category-wise sum of sales in Indonesia:")
print(category_sales_sum)
```

```python
import pandas as pd
import matplotlib.pyplot as plt



# Group by 'Product Name' and calculate the total profit
product_profit = df.groupby('Product Name')['Profit'].sum().reset_index()

# Sort the results by total profit in descending order
product_profit = product_profit.sort_values(by='Profit', ascending=False)

# Plotting
plt.figure(figsize=(12, 6))  # Set the size of the figure
plt.barh(product_profit['Product Name'].head(10), product_profit['Profit'].head(10), color='skyblue')  # Horizontal bar plot
plt.xlabel('Total Profit')
plt.title('Top 10 Products by Total Profit')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest profit on top
plt.grid(axis='x')

# Show the plot
plt.show()
```

```python
# Group by 'Country' and calculate the total profit
country_profit = df.groupby('Country')['Sales'].sum().reset_index()

# Sort the results by total profit in descending order and select the top 10 countries
top_10_countries = country_profit.sort_values(by='Sales', ascending=False).head(10)

# Plotting a bar plot
plt.figure(figsize=(12, 6))  # Set the size of the figure
plt.bar(top_10_countries['Country'], top_10_countries['Sales'], color='skyblue')

# Add titles and labels
plt.xlabel('Country')
plt.ylabel('Sales')
plt.title('Top 10 Countries with Highest Total Sales')
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.grid(axis='y')

# Show the plot
plt.show()
```
